```python
Etching = pd.read_csv('/home/piai/바탕화면/빅데이터프로젝트/Etching_remake.csv').iloc[:,1:];Etching
```

```python
Ion = pd.read_csv('/home/piai/바탕화면/빅데이터프로젝트/Ion_remake.csv').iloc[:,1:];Ion
```

```python
QI = pd.read_csv('/home/piai/바탕화면/빅데이터프로젝트/QualityInspection.csv').iloc[:,1:]; QI
```

```python
Oxid = pd.merge(Oxid,pd.DataFrame(Oxid.Ox_Chamber.value_counts().reset_index()).rename({'index' : 'Ox_Chamber', 'Ox_Chamber' : 'count_oxid'}, axis=1));Oxid.head()
```

```python
soft = pd.merge(soft,pd.DataFrame(soft.photo_soft_Chamber.value_counts().reset_index()).rename({'index' : 'photo_soft_Chamber', 'photo_soft_Chamber' : 'count_soft'}, axis=1));soft.head()
```

```python
litho = pd.merge(litho,pd.DataFrame(litho.lithography_Chamber.value_counts().reset_index()).rename({'index' : 'lithography_Chamber', 'lithography_Chamber' : 'count_litho'}, axis=1));litho.head()
```

```python
Etching = pd.merge(Etching,pd.DataFrame(Etching.Etching_Chamber.value_counts().reset_index()).rename({'index' : 'Etching_Chamber', 'Etching_Chamber' : 'count_etching'}, axis=1));Etching.head()
```

```python
Ion = pd.merge(Ion,pd.DataFrame(Ion.Chamber_Num.value_counts().reset_index()).rename({'index' : 'Chamber_Num', 'Chamber_Num' : 'count_ion'}, axis=1));Ion.head()
```

```python
total_df = pd.merge(Oxid, soft, on=['No_Die', 'Lot_Num', 'Wafer_Num', 'Datetime'])
total_df = pd.merge(total_df, litho, on=['No_Die', 'Lot_Num', 'Wafer_Num', 'Datetime'])
total_df = pd.merge(total_df, Etching, on=['No_Die', 'Lot_Num', 'Wafer_Num', 'Datetime'])
total_df = pd.merge(total_df, Ion, on=['No_Die', 'Lot_Num', 'Wafer_Num', 'Datetime'])
total_df = pd.merge(total_df, QI, on=['No_Die', 'Lot_Num', 'Wafer_Num', 'Datetime'])
```

```python
total_df
```

```python
## 챔버별 (지나간) 웨이퍼 수 합
total_df['chamber_total_count'] = total_df['count_oxid'] + total_df['count_soft'] + total_df['count_litho'] + total_df['count_etching'] + total_df['count_ion']
```

```python
total_df = total_df.astype({'Ox_Chamber' : 'str',
                   'photo_soft_Chamber' : 'str',
                   'lithography_Chamber' : 'str',
                   'Etching_Chamber' : 'str',
                   'Chamber_Num' : 'str'})
```

```python
total_df['chamber_path'] = total_df.Ox_Chamber + total_df.photo_soft_Chamber + total_df.lithography_Chamber + total_df.Etching_Chamber + total_df.Chamber_Num
```

```python
# 불량이면 1 아니면 0
total_df['defective'] = total_df['Target'].apply(lambda x: '1' if x> 194 else '0')
```
