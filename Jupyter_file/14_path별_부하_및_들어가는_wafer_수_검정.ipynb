```python
temp_cnt_chamber_path = df[['No_Die', 'chamber_total_count', 'chamber_path','Target','defective_chip_rate']];temp_cnt_chamber_path
```

```python
temp_df = temp_cnt_chamber_path.groupby(['chamber_path','chamber_total_count'])['defective_chip_rate','Target'].mean().reset_index();temp_df
```

# chi2검정

```python
temp_df = temp_df.sort_values(by = 'chamber_total_count');temp_df
```

```python
small_cnt_path_list  = list(temp_df.head(5).chamber_path.values)
many_cnt_path_list  = list(temp_df.tail(5).chamber_path.values)
```

```python
small_cnt_path_df = temp_df.query('chamber_path in @small_cnt_path_list');small_cnt_path_df
```

```python
many_cnt_path_df = temp_df.query('chamber_path in @many_cnt_path_list');many_cnt_path_df
```

```python
aaa = pd.DataFrame({'0': [small_cnt_path_df.Target.mean(), 2000-small_cnt_path_df.Target.mean()], '1' : [many_cnt_path_df.Target.mean(),2000-many_cnt_path_df.Target.mean()]})

chi,pval,dof,expected = stats.chi2_contingency(aaa.T)

print('chi-square test')
print('chi-sq: {0:0.3f}'.format(chi))
print('p: {0:0.3f}'.format(pval))
```
